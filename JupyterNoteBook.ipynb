```python
## Practice Examples 
## Determine if a year is a leap year using nested "if" statements

year= int(input("Enter a year: "))
if (year % 4 == 0):
    if (year % 100 == 0):
        if (year % 400 == 0):
            print(year, "is a leap year.")
        else:
            print(year, "is not a leap year.")
    else:
        print(year, "is a leap year.")
        
        
else:
    print(year, "is not a leap year.")    
    
        
```

```python
## Loops Control Statements [with Examples]

## Break Statement
## The break statement is used to exit a loop prematurely when a certain condition is met.

for i in range(10):
    if i == 5:
        break  # This will exit the loop when i is equal to 5
    print(i)  # This will print numbers from 0 to 4
```

```python
## Continue Statement
## The continue statement is used to skip the current iteration of a loop and move to the next iteration.

for i in range(10):
    if i % 2 == 0:
        continue  # This will skip even numbers
    print(i)  # This will print odd numbers from 0 to 9
```

```python
## Pass Statement
## The pass statement is a null operation; it does nothing when executed. It is used as a placeholder in loops or functions.
for i in range(10):
    if i==5:
        print("this is I")
        pass  # This will do nothing for numbers less than 5
    print(i)  # This will print numbers from 0 to 9
```

```python
for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Examples:- Prime Numbers between 1 and 100 using a while loop and for loop

for num in range(1, 101):
    if num>1:
        for i in range(2, num):
            if (num%i)==0:
                break
        else:
            print(num, "is a prime number")
```

Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference and symmetric difference.

# Function Examples

Example 1: Password Strength Checker

```python
def is_strong_password(password):
    """this function checks if the password is strong or not"""
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in "!@#$%^()-+" for char in password):
        return False
    return True

## calling the function
print(is_strong_password("WeakPwd"))
print(is_strong_password("Str0ngPwd!"))     ## [str]=String ## [0]=bollan value
      
        
```

Example 2: Calculate the Total Cost of items in a Shopping Cart

```python
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item["price"]* item["quantity"]
        
    return total_cost

## Example cart data

cart=[
    {"name":"Apple","price":5,"quantity":4},
    {"name":"Banana","price":6,"quantity":1},
    {"name":"Orange","price":4,"quantity":2}
    
]

## Calling the Function
total_cost=calculate_total_cost(cart)
print(total_cost)
```

Example 3: Check IF a String Palindrome

```python
## 'Aba' I is both side front and back are same meaning that is palindrome [with Example] 

def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]


print(is_palindrome("Man Become a superman"))  # Should print False, not a palindrome
print(is_palindrome("hello"))  # False, not a palindrome
```

```python
## Factorials = let example [5]= 5x4x3x2x1x0x1.....repeating the loop

def factorial(n):
    if n==0:
        return 1
    else:
        return n * factorial(n-1)
    
print(factorial(5))
```
